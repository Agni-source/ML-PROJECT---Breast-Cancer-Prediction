```python
df=df.drop(columns=['Unnamed: 0','mean fractal dimension','texture error','smoothness error','symmetry error'])
df
```

```python
x =df.drop(['target'], axis = 1) 
y=df['target']
```

```python
# split dataset into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state= 5)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
#feature scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.fit_transform(x_test)
```

```python
#building model

# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)
```

```python
#providing scaled data

lr.fit(x_train_sc, y_train)
```

```python
lr.score(x_test_sc,y_test)
```

```python
# Support vector classifier

from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
```

```python
svc.score(x_test,y_test)
```

```python
#providing scaled data

svc.fit(x_train_sc, y_train)
svc.score(x_test_sc,y_test)
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion = 'entropy', random_state = 51)
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```

```python
#providing scaled data

dt.fit(x_train_sc, y_train)
dt.score(x_test_sc,y_test)
```

```python
#random forest

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(criterion = 'entropy', random_state = 51)
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```

```python
#providing scaled data

rf.fit(x_train_sc, y_train)
rf.score(x_test_sc,y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
adb=AdaBoostClassifier()
adb.fit(x_train,y_train)
```

```python
adb.score(x_test,y_test)
```

```python
#providing scaled data

adb.fit(x_train_sc, y_train)
adb.score(x_test_sc,y_test)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train,y_train)
```

```python
xgb.score(x_test,y_test)
```

```python
#providing scaled data

xgb.fit(x_train_sc, y_train)
xgb.score(x_test_sc,y_test)
```

```python
## Pickle
import pickle

# save model
pickle.dump(xgb,open('breast_cancer_detector.pickle', 'wb'))

#best suitable algorithm
print('Score of XGBoost Algorithm = 98%')
```
